Ploting the graph

```python
import matplotlib.pyplot as plt

st["avg_players"].hist(bins=50)
plt.title("Average players")
plt.show()
```

```python
top_game = st.groupby("name")["avg_players"].mean().sort_values(ascending=False).head(10)
```

```python
top_game
```

```python
top_game.plot(kind='bar', figsize=(10,5), color="red")
plt.title("Top games by average players")
plt.ylabel("Average players")
plt.xlabel("Game name")
plt.xticks(rotation=45)
plt.show()
```
